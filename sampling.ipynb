## Preprocessing Dataset

```python
#concate semua dataset
for i in range(len(lst_data)):
  data = pd.read_csv(lst_data[i])
  data['Tahun'] = lst_data[i]
  data['Tahun'] = data['Tahun'].apply(lambda st: st[st.find("Tahun-")+6:st.find("-Bulan")]).astype(int)
  data['Bulan'] = lst_data[i]
  data['Bulan'] = data['Bulan'].apply(lambda st: st[st.find("Bulan-")+6:st.find(".csv")])
  df_base = pd.concat([df_base, data])
  i+1
```

```python
# fungsi membersihkan bulan
def processkata(kata):
    replace_jan = ['Januari']
    kata = re.sub(r'|'.join(map(re.escape, replace_jan)), '1', kata)
    replace_feb = ['Februari']
    kata = re.sub(r'|'.join(map(re.escape, replace_feb)), '2', kata)
    replace_maret = ['Maret']
    kata = re.sub(r'|'.join(map(re.escape, replace_maret)), '3', kata)
    replace_aprl = ['April']
    kata = re.sub(r'|'.join(map(re.escape, replace_aprl)), '4', kata)
    replace_mei = ['Mei']
    kata = re.sub(r'|'.join(map(re.escape, replace_mei)), '5', kata)
    replace_jun = ['Juni']
    kata = re.sub(r'|'.join(map(re.escape, replace_jun)), '6', kata)
    replace_jul = ['Juli']
    kata = re.sub(r'|'.join(map(re.escape, replace_jul)), '7', kata)
    replace_ags = ['Agustus']
    kata = re.sub(r'|'.join(map(re.escape, replace_ags)), '8', kata)
    replace_sep = ['September']
    kata = re.sub(r'|'.join(map(re.escape, replace_sep)), '9', kata)
    replace_okt = ['Oktober']
    kata = re.sub(r'|'.join(map(re.escape, replace_okt)), '10', kata)
    replace_nov = ['November']
    kata = re.sub(r'|'.join(map(re.escape, replace_nov)), '11', kata)
    replace_des = ['Desember']
    kata = re.sub(r'|'.join(map(re.escape, replace_des)), '12', kata)
    return kata
```

```python
#implementasi fungsi sebelumnya
df_base['Bulan'] = df_base['Bulan'].apply(processkata).astype(int)
```

## Estimasi rata-rata jumlah jiwa terdampak akibat banjir per kelurahan di DKI Jakarta

```python
df1.head()
```

Data banjir yang sudah di lakukan cleaning df1 kemudian akan digunakan untuk menaksir jumlah terdampak jiwa akibat banjir per kelurahan


*   `nama_kecamatan` = Nama kecamatan di Provinsi DKI Jakarta
*   `nama_kelurahan` = Nama kelurahan di Provinsi DKI Jakarta
*   `jumlah_terdampak_jiwa` = Jumlah terdampak jiwa di tiap kelurahan akibat banjir

Diketahui:
* Sejumlah 10 kecamatan akan diambil sebagai PSU

```python
# Summary variables
N = 47              # jumlah seluruh kecamatan di DKI Jakarta
n = 10              # jumlah kecamatan yang disampel
pop_mean = np.mean(data['jumlah_terdampak_jiwa'])

print(f"Jumlah kecamatan di DKI Jakarta                               : {N} kecamatan")
print(f"Jumlah kecamatan yang disampel                                : {n} kecamatan")
print(f'Rata-rata jumlah terdampak jiwa per kelurahan di DKI Jakarta  : {pop_mean:.0f} orang  ')
```

```python
# Random sample 10 kecamatan
unique_kecamatan = df1.groupby('kecamatan', group_keys=False).apply(lambda df1: df1.sample(1))
sample_kecamatan = unique_kecamatan['kecamatan'].sample(n = 10, random_state = 42)
sample_kecamatan
```

```python
# 10 Kecamatan terpilih
kecamatan = df1['kecamatan']

jumlah_kelurahan = df1[kecamatan.isin(['MAMPANG PRAPATAN',
 'SAWAH BESAR',
 'MAMPANG',
 'TAMBORA',
 'KRAMAT JATI',
 'PESANGGRAHAN',
 'JATINEGARA',
 'KEMAYORAN',
 'CILIINCING',
 'MAKASAR'])].groupby('kecamatan')['kelurahan'].count()\
.reset_index(name = 'jumlah_kelurahan')

korban = df1[kecamatan.isin(['MAMPANG PRAPATAN',
 'SAWAH BESAR',
 'MAMPANG',
 'TAMBORA',
 'KRAMAT JATI',
 'PESANGGRAHAN',
 'JATINEGARA',
 'KEMAYORAN',
 'CILIINCING',
 'MAKASAR'])].groupby('kecamatan')['jumlah_terdampak_jiwa'].sum()\
.reset_index(name = 'jumlah_terdampak_jiwa')

kecamatan_terpilih = pd.merge(jumlah_kelurahan, korban, on=['kecamatan'])
kecamatan_terpilih
```

```python
# Mencari nilai mu_r
# y_total adalah jumlah terdampak jiwa dari kecamatan yang disample
y_total = np.sum(kecamatan_terpilih['jumlah_terdampak_jiwa'])

# M adalah jumlah kelurahan dari kecamatan yang disample
M = np.sum(kecamatan_terpilih['jumlah_kelurahan'])

mu_r = y_total / M

print(f'Total jumlah terdampak jiwa dari seluruh kecamatan yang disampel  : {y_total} orang')
print(f'Total jumlah kelurahan dari seluruh kecamatan yang disampel       : {M} kelurahan')
print(f'Estimasi rata-rata jumlah terdampak jiwa                          : {mu_r:.0f} orang')
```

## Estimasi 95% Confidence Interval

```python
# hitung sample varians terlebih dahulu
sc_square = np.sum((kecamatan_terpilih["jumlah_terdampak_jiwa"] - mu_r*kecamatan_terpilih["jumlah_kelurahan"])**2) / (n-1)
print(f"Varians sampel                        : ({np.sqrt(sc_square):.0f})^2")

# selanjutnya hitung estimator varians total populasi
var_y_total_est = N*(N-n)*(sc_square/n)
print(f'Varians dari estimator total populasi : ({np.sqrt(var_y_total_est):.0f}^2)')
```

```python
# cari rata-rata kelurahan dari tiap keamatan yang sample
M_bar = np.mean(kecamatan_terpilih["jumlah_kelurahan"])
print(f"Rata-rata jumlah kelurahan per kecamatan                : {M_bar:.0f}")


# hitung total dari populasi berdasarkan rata-rata
M_tot_est = N * M_bar
print(f"Estimasi jumlah kelurahan dari kecamatan yang disample  : {M_tot_est:.0f}")


# hitung varians dari rata-rata
var_y_mean_est = var_y_total_est / (M_tot_est**2)
print(f"The variance of mean ssu. est.                          : ({np.sqrt(var_y_mean_est):.0f})^2")
```

```python
# hitung z_stat @ 95% CI
alpha = 0.05
z_stat = stats.norm.ppf(1 - alpha/2.)
print(f"z_stat              : {z_stat:.2f}")

# hitung margin of error (d)
d = z_stat * np.sqrt(var_y_mean_est)
print(f"margin of error (d) : {d:.0f}")
```

```python
# hitung confidence interval
lb = mu_r - d
ub = mu_r + d

print(f'Dengan 95% confidence interval, nilai rata-rata asli jumlah terdampak jiwa per kelurahan di DKI Jakarta\
berkisar antara {lb:.0f} sampai {ub:.0f} orang')
```
